```python
import pygame
import time
import random
 
pygame.init()
 
white = (255, 255, 255)
black = (0, 0, 0)
red = (213, 50, 80)
green = (0, 255, 0)
gray = (180,180,180)
 
dis_width = 600
dis_height = 600
 
dis = pygame.display.set_mode((dis_width, dis_height))
pygame.display.set_caption('Snake powered by Smirnov')
 
clock = pygame.time.Clock()
 
snake = 10
speed = 15
 
font_style = pygame.font.SysFont("impact", 25)
score_font = pygame.font.SysFont("arial", 35)
 
def our_snake(snake, snake_list):
    for x in snake_list:
        pygame.draw.rect(dis, black, [x[0], x[1], snake, snake])
 
 
def length(score):
    value = score_font.render("Length: " + str(score), True, white)
    dis.blit(value, [8, 3])
 
 
def message(msg, color):
    mesg = font_style.render(msg, True, color)
    dis.blit(mesg, [80, 300])
 
 
def gameLoop():
    game_over = False
    game_close = False
 
    x1 = dis_width / 2
    y1 = dis_height / 2
 
    x1_change = 0
    y1_change = 0
 
    snake_List = []
    snake_len = 1
 
    foodx = round(random.randrange(0, dis_width - snake) / 10.0) * 10.0
    foody = round(random.randrange(0, dis_height - snake) / 10.0) * 10.0
 
    while not game_over:
 
        while game_close == True:
            dis.fill(gray)
            message("You Lost! Press K to Play Again or M to Quit", red)
            length(snake_len)
            pygame.display.update()
 
            for event in pygame.event.get():
                if event.type == pygame.KEYDOWN:
                    if event.key == pygame.K_m:
                        game_over = True
                        game_close = False
                    if event.key == pygame.K_k:
                        gameLoop()
 
        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                game_over = True
            if event.type == pygame.KEYDOWN:
                if event.key == pygame.K_UP:
                    y1_change = -snake
                    x1_change = 0
                elif event.key == pygame.K_DOWN:
                    y1_change = snake
                    x1_change = 0
                elif event.key == pygame.K_RIGHT:
                    x1_change = snake
                    y1_change = 0
                elif event.key == pygame.K_LEFT:
                    x1_change = -snake
                    y1_change = 0

 
        if x1 >= dis_width or x1 < 0 or y1 >= dis_height or y1 < 0:
            game_close = True
        x1 += x1_change
        y1 += y1_change
        dis.fill(gray)
        pygame.draw.rect(dis, red, [foodx, foody, snake, snake])
        snake_Head = []
        snake_Head.append(x1)
        snake_Head.append(y1)
        snake_List.append(snake_Head)
        if len(snake_List) > snake_len:
            del snake_List[0]
 
        for x in snake_List[:-1]:
            if x == snake_Head:
                game_close = True
 
        length(snake_len)
        our_snake(snake, snake_List)
 
        pygame.display.update()
 
        if x1 == foodx and y1 == foody:
            foodx = round(random.randrange(0, dis_width - snake) / 10.0) * 10.0
            foody = round(random.randrange(0, dis_height - snake) / 10.0) * 10.0
            snake_len += 1
 
        clock.tick(speed)
 
    pygame.quit()
    quit()
    
      
gameLoop()
```
